```python
plt.figure(figsize=(20,20))
plotnumber = 1

for column in df2.drop(columns = ['3 Which city do you shop online from?']):
    if plotnumber <= 40:
        ax = plt.subplot(10,4,plotnumber)
        sns.distplot(df2[column])
        plt.xlabel(column)
    
    plotnumber+=1
plt.tight_layout()
```

```python
plt.figure(figsize=(20,20))
plotnumber = 1

for column in df2.drop(columns = ['3 Which city do you shop online from?']):
    if plotnumber <= 40:
        ax = plt.subplot(10,4,plotnumber)
        sns.boxplot(df2[column])
        plt.xlabel(column)
    
    plotnumber+=1
plt.tight_layout()
```

    These are the Normal Distribution & Outliers of the dataset. And almost all the columns are categorical columns as they are encoded ones. So we will be getting these kind of plots.
